```python
# crear una nueva columna con el nombre de "Tesla" y en la posición 1 para que aparezca en cada fila su acrónimo "TSLA"
tesla.insert(1, "Tesla", "TSLA")
```

```python
# Cambiar el nombre de las columnas "último", "Vol." y "% var." para mayor comodidad
tesla = tesla.rename(columns={"Último": "Cierre", "Vol.": "Volumen.M", "% var.": "Var.%"})
```

```python
# Reemplazamos de las columnas "Vol." y de la columna "%var. Los strings de "M" y "%" respectivamente
tesla['Volumen.M'] = tesla['Volumen.M'].replace('M', '', regex=True)
tesla["Var.%"] = tesla["Var.%"].replace("%", "", regex= True )
```

```python
# cambiamos de tipo los datos de strings a tipo float para poder hacer los cálculos usando astipe(float)
# pero antes debemos reemplazar las "," por los "." con replace() y usando lambda para una mejor practicidad
columnas_numericas = ["Cierre", "Apertura", "Máximo", "Mínimo", "Volumen.M", "Var.%" ]
tesla[columnas_numericas] = tesla[columnas_numericas].apply(lambda x: x.str.replace(',', '.'))
tesla[columnas_numericas] = tesla[columnas_numericas].astype(float)
```

```python
# elimino la columna de Volumen
del tesla["Volumen.M"]
```

```python
tesla.head()
```

```python
# Cambiamos la columna 'Fecha' de tipo string a tipo datetime
tesla['Fecha'] = pd.to_datetime(tesla['Fecha'])
```

```python
tesla.dtypes
```

```python
# Invertir el orden de las filas para que el primer valor sea de la fecha más antigua y el último valor de la fecha más reciente
tesla = tesla.iloc[::-1].reset_index(drop=True)
tesla.head()
```

```python
# crear una nueva columna con el nombre de "Apple" y en la posición 1 para que aparezca en cada fila su acrónimo "APPL"
apple.insert(1, "Apple", "APPL")

# Cambiar el nombre de las columnas "último", "Vol." y "% var." para mayor comodidad
apple = apple.rename(columns={"Último": "Cierre", "Vol.": "Volumen.M", "% var.": "Var.%"})

# Reemplazamos de las columnas "Vol." y de la columna "%var. Los strings de "M" y "%" respectivamente
apple['Volumen.M'] = apple['Volumen.M'].replace('M', '', regex=True)
apple['Volumen.M'] = apple['Volumen.M'].replace('B', '', regex=True)
apple["Var.%"] = apple["Var.%"].replace("%", "", regex= True )

# cambiamos de tipo los datos de strings a tipo float para poder hacer los cálculos usando astupe()
# pero antes debemos reemplazar las "," por los "." con replace() y usando lambda para una mejor parcticidad
columnas_numericas = ["Cierre", "Apertura", "Máximo", "Mínimo", "Volumen.M", "Var.%" ]
apple[columnas_numericas] = apple[columnas_numericas].apply(lambda x: x.str.replace(',', '.'))
apple[columnas_numericas] = apple[columnas_numericas].astype(float)
```

### DATAFRAMES UNIÓN

```python
# cuadrar fechas de todos los dataframes, estableciendo el rango de fechas común
fecha_inicio = '2023-05-19'
fecha_fin = '2010-07-19'
rango_fechas = pd.date_range(start=fecha_inicio, end=fecha_fin, freq='D') # Frecuencia diaria
```

```python
# Para establecer el rango de fechas a cada DataFrame
tesla = pd.merge(tesla, pd.DataFrame({'Fecha': rango_fechas}), how='outer', on='Fecha')
apple = pd.merge(apple, pd.DataFrame({'Fecha': rango_fechas}), how='outer', on='Fecha')
microsoft = pd.merge(microsoft, pd.DataFrame({'Fecha': rango_fechas}), how='outer', on='Fecha')
oro = pd.merge(apple, pd.DataFrame({'Fecha': rango_fechas}), how='outer', on='Fecha')
bitcoin = pd.merge(apple, pd.DataFrame({'Fecha': rango_fechas}), how='outer', on='Fecha')
nasdaq = pd.merge(apple, pd.DataFrame({'Fecha': rango_fechas}), how='outer', on='Fecha')

```

```python
# Comprobamos que las filas no terminan de cuadrar, ya que hay 1 en tesla que no cuadra
print("Tesla tiene", tesla.shape[0], "filas y", tesla.shape[1], "columnas")
print("--" *30)
print("Apple tiene", apple.shape[0], "filas y", apple.shape[1], "columnas")
print("--" *30)
print("Microsoft tiene", microsoft.shape[0], "filas y", microsoft.shape[1], "columnas")
print("--" *30)
print("Oro tiene", oro.shape[0], "filas y", oro.shape[1], "columnas")
print("--" *30)
print("Bitcoin tiene", bitcoin.shape[0], "filas y", bitcoin.shape[1], "columnas")
print("--" *30)
print("Nasdaq tiene", nasdaq.shape[0], "filas y", nasdaq.shape[1], "columnas")
```

```python
# Unión de todos los DataFrames por Fecha común e incorporando sufijos para aclarar el nombre de las columnas
merged = pd.merge(tesla, apple, on='Fecha', how='outer', suffixes=('_Tesla', '_Apple'))
merged2 = pd.merge(merged, microsoft, on='Fecha', how='outer', suffixes=('_Apple', ''))
merged3 = pd.merge(merged2, oro, on='Fecha', how='outer', suffixes=('_Microsoft', '_Oro'))
merged4 = pd.merge(merged3, bitcoin, on='Fecha', how='outer', suffixes=('_oro', '_Bitcoin'))
activos_financieros = pd.merge(merged4, nasdaq, on='Fecha', how='outer', suffixes=('Bitcoin', '_Nasdaq'))
activos_financieros.tail()
```
