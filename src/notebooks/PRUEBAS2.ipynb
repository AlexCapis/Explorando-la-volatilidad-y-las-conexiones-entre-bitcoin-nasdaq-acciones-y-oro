```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Cargar los datos en un DataFrame
data = pd.read_csv('all_data.csv')
```

```python


# Calcular los retornos diarios
data['Retorno_diario_Bitcoin'] = data['Cierre_Bitcoin'].pct_change()
data['Retorno_diario_Nasdaq'] = data['Cierre_Nasdaq'].pct_change()
data['Retorno_diario_Tesla'] = data['Cierre'].pct_change()
data['Retorno_diario_Apple'] = data['Cierre_Apple'].pct_change()
data['Retorno_diario_Oro'] = data['Cierre_Oro'].pct_change()
data['Retorno_diario_Microsoft'] = data['Cierre_Microsoft'].pct_change()

# Calcular la volatilidad diaria
volatility = data[['Retorno_diario_Bitcoin', 'Retorno_diario_Nasdaq', 'Retorno_diario_Tesla', 'Retorno_diario_Apple', 'Retorno_diario_Oro', 'Retorno_diario_Microsoft']].std()

# Mostrar la volatilidad
print('Volatilidad Diaria:')
print(volatility)

# Generar un gráfico de barras para visualizar la volatilidad
volatility.plot(kind='bar', figsize=(10, 6))
plt.title('Volatilidad Diaria')
plt.xlabel('Activos Financieros')
plt.ylabel('Volatilidad')
plt.show()

```
